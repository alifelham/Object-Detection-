## LOADING MODEL

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

## DETECTION

```python
img = 'https://images.unsplash.com/photo-1589828155685-83225f7d91f3?ixlib=rb-1.2.1&ixid=MnwxMjA3fDB8MHxzZWFyY2h8Nnx8Y2FyJTIwcm9hZHxlbnwwfHwwfHw%3D&w=1000&q=80'
```

traffic.jpeg

```python
results = model(img)
results.print()
```

The model detects 3 persons, 26 cars, 1 motorcycle, 1 bus, and 2 trucks from the given image.

```python
%matplotlib inline
from matplotlib.pyplot import figure
figure(figsize = (20, 15))
plt.imshow(np.squeeze(results.render()))
plt.show()
```
